## Summarize Metrics

```python
import numpy as np
import cv2
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
# Resnet special case where tp->fn
# Other neural network models: tp->tn, fp->fn 


def calculate_pixel_accuracy(true_positives,
                    true_negatives,
                    false_positives,
                    false_negatives):
    return (true_positives+true_negatives)/(true_positives+true_negatives+false_positives+false_negatives)
    
    
def calculate_precision(true_positives,
                    true_negatives,
                    false_positives,
                    false_negatives):
    return (true_positives)/(true_positives+false_positives)
    
    
def calculate_recall(true_positives,
                    true_negatives,
                    false_positives,
                    false_negatives):
    return (true_positives)/(true_positives+false_negatives)


def calculate_jaccard_binary(true_positives,
                    true_negatives,
                    false_positives,
                    false_negatives):
    return (true_positives)/(true_positives+false_positives+false_negatives)


    
def calculate_fpr_instance(tn, fp):
    return 1.0 - (tn/(tn+fp))


def calculate_tpr_instance(tp, fn):
    return tp/(tp+fn)


def calculate_auc(true_positives,
                    true_negatives,
                    false_positives,
                    false_negatives):
    fpr_instance = calculate_fpr_instance(true_negatives, false_positives)
    tpr_instance = calculate_tpr_instance(true_positives, false_negatives)
    a1 = (tpr_instance*fpr_instance)/2.0
    a2 = ((1-fpr_instance)*(1-tpr_instance))/2.0
    a3 = (1-fpr_instance)*tpr_instance
    return a1+a2+a3

    

    
def display_confusion_matrix(name,
                    true_positives,
                    true_negatives,
                    false_positives,
                    false_negatives):
    cm = np.array([
        [true_negatives, false_positives],
        [false_negatives, true_positives]])
    ls = [0, 1] # your y labels
    disp = ConfusionMatrixDisplay(confusion_matrix=np.array(cm), display_labels=ls)
    fig, ax = plt.subplots(figsize=(2,2))
    ax.set_title(name, fontsize=8)
    disp.plot(ax=ax, colorbar=True, cmap='plasma')
    
    
def display_auc_roc(name,
                    true_positives,
                    true_negatives,
                    false_positives,
                    false_negatives):
    fpr_instance = calculate_fpr_instance(true_negatives, false_positives) #1.0 - (true_negatives/(true_negatives+false_positives))
    tpr_instance = calculate_tpr_instance(true_positives, false_negatives) #true_positives/(true_positives+false_negatives)
    
    fpr = [0.0, fpr_instance, 1.0]
    tpr = [0.0, tpr_instance, 1.0]
    
    fig, ax = plt.subplots()
    ax.plot(fpr, tpr, linewidth=2.0)
    plt.show()


    
def display_metrics(name,
                true_positives,
                true_negatives,
                false_positives,
                false_negatives):
    pixel_accuracy = calculate_pixel_accuracy(true_positives, true_negatives, false_positives, false_negatives)
    precision = calculate_precision(true_positives, true_negatives, false_positives, false_negatives)
    recall = calculate_recall(true_positives, true_negatives, false_positives, false_negatives)
    jaccard = calculate_jaccard_binary(true_positives, true_negatives, false_positives, false_negatives)
    auc = calculate_auc(true_positives, true_negatives, false_positives, false_negatives)
    
    print("name: {}".format(name))
    print("true positive: {}".format(true_positives))
    print("true negative: {}".format(true_negatives))
    print("false positive: {}".format(false_positives))
    print("false negative: {}".format(false_negatives))
    print("pixel accuracy: {}".format(pixel_accuracy))
    print("recall: {}".format(recall))
    print("jaccard: {}".format(jaccard))
    print("auc: {}".format(auc))
    print("precision: {}".format(precision))

    print()

    display_confusion_matrix(name, true_positives, true_negatives, false_positives, false_negatives)
    #display_auc_roc(name, true_positives, true_negatives, false_positives, false_negatives)
    
    



    
    
    
def plot_roc_curves(
    logistic_regression_tp_tn_fp_fn, 
    svm_tp_tn_fp_fn,
    random_forest_tp_tn_fp_fn,
    resnet_tp_tn_fp_fn,
    mobilenet_v2_tp_tn_fp_fn,
    mobilenet_etpu_tp_tn_fp_fn
):
    logistic_regression_fpr = [0.0, calculate_fpr_instance(logistic_regression_tp_tn_fp_fn[1], logistic_regression_tp_tn_fp_fn[2]), 1.0]
    logistic_regression_tpr = [0.0, calculate_tpr_instance(logistic_regression_tp_tn_fp_fn[0], logistic_regression_tp_tn_fp_fn[3]), 1.0]
    
    svm_fpr = [0.0, calculate_fpr_instance(svm_tp_tn_fp_fn[1], svm_tp_tn_fp_fn[2]), 1.0]
    svm_tpr = [0.0, calculate_tpr_instance(svm_tp_tn_fp_fn[0], svm_tp_tn_fp_fn[3]), 1.0]
    
    random_forest_fpr = [0.0, calculate_fpr_instance(random_forest_tp_tn_fp_fn[1], random_forest_tp_tn_fp_fn[2]), 1.0]
    random_forest_tpr = [0.0, calculate_tpr_instance(random_forest_tp_tn_fp_fn[0], random_forest_tp_tn_fp_fn[3]), 1.0]
    
    resnet_fpr = [0.0, calculate_fpr_instance(resnet_tp_tn_fp_fn[1], resnet_tp_tn_fp_fn[2]), 1.0]
    resnet_tpr = [0.0, calculate_tpr_instance(resnet_tp_tn_fp_fn[0], resnet_tp_tn_fp_fn[3]), 1.0]
    
    mobilenet_v2_fpr = [0.0, calculate_fpr_instance(mobilenet_v2_tp_tn_fp_fn[1], mobilenet_v2_tp_tn_fp_fn[2]), 1.0]
    mobilenet_v2_tpr = [0.0, calculate_tpr_instance(mobilenet_v2_tp_tn_fp_fn[0], mobilenet_v2_tp_tn_fp_fn[3]), 1.0]
    
    
    mobilenet_etpu_fpr = [0.0, calculate_fpr_instance(mobilenet_etpu_tp_tn_fp_fn[1], mobilenet_etpu_tp_tn_fp_fn[2]), 1.0]
    mobilenet_etpu_tpr = [0.0, calculate_tpr_instance(mobilenet_etpu_tp_tn_fp_fn[0], mobilenet_etpu_tp_tn_fp_fn[3]), 1.0]
    
    
    fig, ax = plt.subplots()
    plt.title("ROCs")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    ax.plot([0.0, 1.0], [0.0, 1.0], '--', color='gray')
    ax.plot(logistic_regression_fpr, logistic_regression_tpr, linewidth=2.0, label='Logistic Regression')
    ax.plot(svm_fpr, svm_tpr, linewidth=2.0, label='Support Vector Machine')
    ax.plot(random_forest_fpr, random_forest_tpr, linewidth=2.0, label='Random Forest')
    ax.plot(resnet_fpr, resnet_tpr, linewidth=2.0, label='ResNetV1')
    ax.plot(mobilenet_v2_fpr, mobilenet_v2_tpr, linewidth=2.0, label='MobileNetV2')
    ax.plot(mobilenet_etpu_fpr, mobilenet_etpu_tpr, linewidth=2.0, label='MobileNetEdgeTPU')
    ax.legend(loc="lower right")
    plt.show()

    


def show_all_metrics(
    logistic_regression_tp_tn_fp_fn, 
    svm_tp_tn_fp_fn,
    random_forest_tp_tn_fp_fn,
    resnet_tp_tn_fp_fn,
    mobilenet_v2_tp_tn_fp_fn,
    mobilenet_etpu_tp_tn_fp_fn):

    display_metrics(name="Logistic Regression",
                true_positives=logistic_regression_tp_tn_fp_fn[0],
                true_negatives=logistic_regression_tp_tn_fp_fn[1],
                false_positives=logistic_regression_tp_tn_fp_fn[2],
                false_negatives=logistic_regression_tp_tn_fp_fn[3])
    
    display_metrics(name="SVM",
                true_positives=svm_tp_tn_fp_fn[0],
                true_negatives=svm_tp_tn_fp_fn[1],
                false_positives=svm_tp_tn_fp_fn[2],
                false_negatives=svm_tp_tn_fp_fn[3])
        
    display_metrics(name="Random Forest",
                true_positives=random_forest_tp_tn_fp_fn[0],
                true_negatives=random_forest_tp_tn_fp_fn[1],
                false_positives=random_forest_tp_tn_fp_fn[2],
                false_negatives=random_forest_tp_tn_fp_fn[3])    
    
    display_metrics(name="ResNetV1",
                true_positives=resnet_tp_tn_fp_fn[0],
                true_negatives=resnet_tp_tn_fp_fn[1],
                false_positives=resnet_tp_tn_fp_fn[2],
                false_negatives=resnet_tp_tn_fp_fn[3])    
    
    display_metrics(name="MobileNetV2",
                true_positives=mobilenet_v2_tp_tn_fp_fn[0],
                true_negatives=mobilenet_v2_tp_tn_fp_fn[1],
                false_positives=mobilenet_v2_tp_tn_fp_fn[2],
                false_negatives=mobilenet_v2_tp_tn_fp_fn[3])    
    
    display_metrics(name="MobileNetEdgeTPU",
                true_positives=mobilenet_etpu_tp_tn_fp_fn[0],
                true_negatives=mobilenet_etpu_tp_tn_fp_fn[1],
                false_positives=mobilenet_etpu_tp_tn_fp_fn[2],
                false_negatives=mobilenet_etpu_tp_tn_fp_fn[3])    
    
        
    plot_roc_curves(
        logistic_regression_tp_tn_fp_fn, 
        svm_tp_tn_fp_fn,
        random_forest_tp_tn_fp_fn,
        resnet_tp_tn_fp_fn,
        mobilenet_v2_tp_tn_fp_fn,
        mobilenet_etpu_tp_tn_fp_fn
    )


```
