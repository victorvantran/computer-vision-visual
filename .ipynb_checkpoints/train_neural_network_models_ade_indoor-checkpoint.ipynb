## Auxiliary Functions 

```python
NON_OBSTACLE_CLASS_LABEL = 0
OBSTACLE_CLASS_LABEL = 1

def save_dataframe(df, dataset_dir, name, index=False):
    ''' Save dataframe '''
    dataframes_dir = os.path.join(dataset_dir, "dataframes")     
    df.to_csv(os.path.join(dataframes_dir, name), index=index)

    
def load_dataframe(filename, dataset_dir, header=None, sep=',', index_col=0):
    ''' Load dataframe '''
    dataframes_dir = os.path.join(dataset_dir, "dataframes")     
    return pd.read_csv(os.path.join(dataframes_dir, filename), header=header, sep=sep, index_col=index_col, lineterminator='\n')


def interpolate_class_id_invalid(label, class_id_invalid):
    ''' Interpolate the label to fill invalid nodes with the nearest neighbor '''
    mask = (label != class_id_invalid)
    if np.count_nonzero(mask) == 0:
        return label # The entire label is invalid node
    xx, yy = np.meshgrid(np.arange(label.shape[1]), np.arange(label.shape[0]))
    xym = np.vstack( (np.ravel(xx[mask]), np.ravel(yy[mask])) ).T

    # Get valid class ids
    valid_class_ids = np.ravel(label[:,:][mask] )
    # Interpolate
    interp = scipy.interpolate.NearestNDInterpolator( xym, valid_class_ids )
    # Reshape to the 2D matrix
    interpolated_label = interp(np.ravel(xx), np.ravel(yy)).reshape( xx.shape )
    interpolated_label = interpolated_label.astype(np.uint16)
    
    # Assert that there are no more invalid pixels
    assert(np.count_nonzero(interpolated_label==0)==0)
    return interpolated_label


def naive_bin(label, walkable_class_ids):
    ''' All walkable classes will be reasigned the id 0 and the rest will be reassigned id 1 '''
    assert (len(walkable_class_ids) > 0)
    walkable = (label == walkable_class_ids[0])
    for walkable_class_id in walkable_class_ids:
        walkable = (walkable | (label == walkable_class_id))

    non_walkable = np.invert(walkable)

    label[walkable] = NON_OBSTACLE_CLASS_LABEL
    label[non_walkable] = OBSTACLE_CLASS_LABEL
    
    return label
```

## Create SegmentationClassRaw 
### Create the Raw Segmentation File for each .png file. The difference between a raw segmentation file and the .png is that the .png holds RGB data representing the color-coded class. The raw segementation file holds just the class id for each pixel. This reduces the data size by a factor of 3.

```python
# We are going to remove the invalid nodes as best we can using nearest neighbor interpolation.
TRAINING_ENVIRONMENT_DIR = r"../training_environment"
ADE_INDOOR_ENVIRONMENT_DIR = os.path.join(TRAINING_ENVIRONMENT_DIR, r"ade_indoor")


DATASETS_DIR = r"../datasets"
ADE_DATASET_DIR = os.path.join(DATASETS_DIR, "ADE")
ADE_MISC_DIR = os.path.join(ADE_DATASET_DIR, "miscellaneous")     
ADE_CLASS_FILE =  os.path.join(ADE_MISC_DIR, r"class_list.txt")

ADE_CLASS_ID_INVALID = 0



def get_classes(class_file):
    ''' Return a list of classes for ADE (should be 150 unqiue classes) '''
    classes_df = pd.read_csv(class_file, header=0, sep='\t' , lineterminator='\n')
    classes = np.asarray(classes_df['Name'])
    for i in range(len(classes)):
        classes[i] = re.split(',', classes[i])[0]
        classes[i] = classes[i].rstrip()
    return classes


def create_segclassraw(training_environment_dir, reference_dataset_dir, df, class_id_invalid):
    ''' Create the raw segmentation class. The non-obstacle class (0) is floor and rug. The obstacle class are the other classes. '''
    segclassraw_dir = os.path.join(training_environment_dir, "SegmentationClassRaw")    
    labels_dir = os.path.join(reference_dataset_dir, "labels") 
    
    for instance in (df.loc[:, ['id', 'dataset', 'invalid']]).itertuples():
        label_file_name = instance.dataset + "_" + str(instance.id).zfill(ZFILL) + "." + "npy"
        label_file = os.path.join(labels_dir, label_file_name) 
        
        seg_raw = np.load(label_file)
        seg_processed = seg_raw
        
        # Create labels_interpolated directory
        if instance.invalid <= 0.1:
            seg_interpolated = interpolate_class_id_invalid(label=seg_raw, class_id_invalid=class_id_invalid)
            seg_processed = seg_interpolated

        # Bin
        misc_dir = os.path.join(reference_dataset_dir, "miscellaneous")     
        class_file =  os.path.join(misc_dir, r"class_list.txt")

        classes = get_classes(class_file=class_file)
        walkable_set = set(["floor", "rug"])
        walkable_class_ids = [(list(classes).index(class_name)+1) for class_name in walkable_set]
    
        seg_processed = naive_bin(label=seg_processed, walkable_class_ids=walkable_class_ids)
        
        # Save
        seg_raw_file_name = instance.dataset + "_" + str(instance.id).zfill(ZFILL) + "." + "png"
        segmentation_class = Image.fromarray(seg_processed.astype(dtype=np.uint8))
        segmentation_class.save(os.path.join(segclassraw_dir, seg_raw_file_name))
        np.save(os.path.join(segclassraw_dir, label_file_name), seg_processed)   
        
        
def get_segraw(dataset, instance_id, training_environment_dir):
    ''' Get a interpolated label from an instance as a npy matrix of shape [height, width] '''
    segclassraw_dir = os.path.join(training_environment_dir, "SegmentationClassRaw")    
    im_frame = Image.open(os.path.join(segclassraw_dir, dataset + "_" + str(instance_id).zfill(8) + "." + "png"))
    np_frame = np.array(im_frame)
    return np_frame


def demonstrate_interpolation():
    seg_interpolated = get_segraw(dataset="ADE", instance_id=0, training_environment_dir=ADE_INDOOR_ENVIRONMENT_DIR)
    print(seg_interpolated)
    plt.imshow(seg_interpolated, cmap='turbo', interpolation='none', vmin=0, vmax=1)
    plt.title("Interpolated Segmentation Image")
    plt.show()


def run_create_segmentation_class_raw():
    %cd G:\segmentation_master\notebooks
    create_segclassraw(
        training_environment_dir=ADE_INDOOR_ENVIRONMENT_DIR, 
        reference_dataset_dir=ADE_DATASET_DIR, 
        df=df_ade_indoor,
        class_id_invalid=ADE_CLASS_ID_INVALID)

    demonstrate_interpolation()


demonstrate_interpolation()
```

## Create List
### Create a randomized list that splits the dataset into train, validation, and test

```python
TRAINING_ENVIRONMENT_DIR = r"../training_environment"
ADE_INDOOR_ENVIRONMENT_DIR = os.path.join(TRAINING_ENVIRONMENT_DIR, r"ade_indoor")

DATASETS_DIR = r"../datasets"
ADE_DATASET_DIR = os.path.join(DATASETS_DIR, "ADE")
ADE_MISC_DIR = os.path.join(ADE_DATASET_DIR, "miscellaneous")     
ADE_CLASS_FILE =  os.path.join(ADE_MISC_DIR, r"class_list.txt")

TEST_SIZE = 0.80


def create_list(df, training_environment_dir, test_size):
    ''' Create a text file listing the randomized, stratified, sampling instances for train, validation, and test'''
    # train_test_split
    df = convert_misc_scene(df)
    train_df, other_df = train_test_split(df, stratify=df['scene'], test_size=1-test_size, shuffle=True, random_state=RANDOM_SEED)

    other_df = convert_misc_scene(other_df)
    val_df, test_df = train_test_split(other_df, stratify=other_df['scene'], test_size=0.5, shuffle=True, random_state=RANDOM_SEED)
    
    train_instance_names = []
    for instance in (train_df.loc[:, ['id', 'dataset']]).itertuples():
        train_instance_names.append("{}_{}".format(instance.dataset, str(int(instance.id)).zfill(ZFILL)))

    val_instance_names = []
    for instance in (val_df.loc[:, ['id', 'dataset']]).itertuples():
        val_instance_names.append("{}_{}".format(instance.dataset, str(int(instance.id)).zfill(ZFILL)))
        
    test_instance_names = []
    for instance in (test_df.loc[:, ['id', 'dataset']]).itertuples():
        test_instance_names.append("{}_{}".format(instance.dataset, str(int(instance.id)).zfill(ZFILL)))
        
    train_instance_names = sorted(train_instance_names)
    val_instance_names = sorted(val_instance_names)
    train_val_instance_names = sorted(train_instance_names + val_instance_names)
    test_instance_names = sorted(test_instance_names)
    
    print("Training instances: {}".format(len(train_instance_names)))
    print("Validation instances: {}".format(len(val_instance_names)))
    print("TrainVal instances: {}".format(len(train_val_instance_names)))
    print("Test instances: {}".format(len(test_instance_names)))

    with open(os.path.join(os.path.join(training_environment_dir, "List"), "train.txt"), mode='wt', encoding='utf-8') as myfile:
        myfile.write('\n'.join(train_instance_names))   

    with open(os.path.join(os.path.join(training_environment_dir, "List"), "val.txt"), mode='wt', encoding='utf-8') as myfile:
        myfile.write('\n'.join(val_instance_names))  
        
    with open(os.path.join(os.path.join(training_environment_dir, "List"), "trainval.txt"), mode='wt', encoding='utf-8') as myfile:
        myfile.write('\n'.join(train_val_instance_names))    

    with open(os.path.join(os.path.join(training_environment_dir, "List"), "test.txt"), mode='wt', encoding='utf-8') as myfile:
        myfile.write('\n'.join(test_instance_names))    
        
        
def run_create_list():
    %cd G:\segmentation_master\notebooks
    create_list(df=df_ade_indoor, training_environment_dir=ADE_INDOOR_ENVIRONMENT_DIR, test_size=TEST_SIZE)
```

## Create Representative Dataset Numpy 
### Create a representatitive dataset from the original dataset (say 256 samples). This dataset is used for post-quantized-aware training

```python
DATASETS_DIR = r"../datasets"
ADE_DATASET_DIR = os.path.join(DATASETS_DIR, "ADE")

TRAINING_ENVIRONMENT_DIR = r"../training_environment"
ADE_INDOOR_ENVIRONMENT_DIR = os.path.join(TRAINING_ENVIRONMENT_DIR, r"ade_indoor")


''' Normalizing the size of our images and respective mask to 512x512 '''
def crop_bottom_center(instance):
    ''' Crop the image to a square by the bottom center '''
    y,x = instance.shape[0], instance.shape[1]
    new_size = min(x, y)
    start_x = x//2-(new_size//2)
    start_y = y-new_size
    return instance[start_y:start_y+new_size, start_x:start_x+new_size]


def normalize_image(image, target_width, target_height):
    ''' We will use Bilinear Interpolation to resize the original image to the target dimension. '''
    # First crop to a square
    cropped_image = crop_bottom_center(image)
    # Resize the square to the target width and height
    normalized_image = cv2.resize(cropped_image, dsize=(target_width, target_height), interpolation=cv2.INTER_LINEAR)
    return normalized_instance

def create_representative_dataset(dataset_dir, training_environment_dir, num_instances, width, height):
    ''' Creates a representative dataset from the test list '''
    list_dir = os.path.join(training_environment_dir, "List")
    list_file = os.path.join(list_dir, "train.txt")
    instances = open(list_file).read().splitlines()
    np.random.seed(RANDOM_SEED)
    sampled_instances = np.random.choice(instances, num_instances)
    #print(sampled_instances)
    
    ade_images_dir = os.path.join(dataset_dir, "images")
    representative_dataset = list()
    for instance_name in sampled_instances:
        #print(os.path.join(ade_instances_dir, instance + "." + "npy"))
        image = np.array(Image.open(os.path.join(ade_images_dir, instance_name + "." + "jpg")))
        normalized_image = np.expand_dims(normalize_instance(image=image, target_width=TARGET_WIDTH, target_height=TARGET_HEIGHT), 0)
        #print(normalized_image.shape)
        #break
        representative_dataset.append(normalized_image)
        
    representative_dataset = np.array(representative_dataset)
    
    ade_miscellaneous_dir = os.path.join(dataset_dir, "miscellaneous")
    np.save(os.path.join(ade_miscellaneous_dir, "representative" + "." + "npy"), representative_dataset)

    return representative_dataset


def run_create_representative_dataset():
    representative_dataset = create_representative_dataset(dataset_dir=ADE_DATASET_DIR, training_environment_dir=ADE_INDOOR_ENVIRONMENT_DIR, 
                                  num_instances=250,
                                  width=257, height=257)
    print(representative_dataset.shape)
```
